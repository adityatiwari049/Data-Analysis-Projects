# How much amount the Companies Spend on R & D

```python
df.head()
```

```python
print("Company Spending on R&D in $Bn")
RD =df.groupby('Company')['R&D_Spending_USD_Mn'].sum()/1000
RD
```

```python
sns.barplot(x=RD.index , y=RD.values , color='Red', edgecolor='black')
#sns.set_theme(style='w')
plt.title("Company R&D Spending" )
plt.xlabel("Company ")
plt.ylabel("Amount in USD_$Bn")
plt.tight_layout()
plt.show()
```

# Revenue Earned by the Companies 


```python
df.head()
```

```python
RE =df.groupby('Company')['AI_Revenue_USD_Mn'].sum()/1000
RE
```

```python
df2 = pd.DataFrame({
    'Company': RE.index,
    'Revenue': RE.values
})
fig = px.bar(
    df2,
    x='Company',
    y='Revenue',
    color='Revenue',
    color_continuous_scale='Viridis',
    hover_data={'Revenue': ':.2f'}  # format revenue with 2 decimals
)
fig.update_layout(
    title='Revenue Earned by Companies in USD_$Bn',
    xaxis_title='Company',
    yaxis_title='Revenue in USD_$Bn'
)
fig.show()
```

```python
# bar plot to show company Revenue and Expenditure by each Company
sns.set_theme(style='darkgrid')
plt.figure(figsize=(11, 5))
plt.subplot(1 , 2 , 1)
plt.bar(RD.index , RE.values , color = ['#a6cee3', '#fdbf6f', '#b2df8a'] )
plt.xlabel("Company ")
plt.ylabel("Amount in USD_$Bn")

plt.subplot(1,2,2)
plt.bar(RE.index , RE.values ,color = ['#a6cee3', '#fdbf6f', '#b2df8a'] )
plt.xlabel('Company')
plt.ylabel('Revenue in USD_$Bn')
plt.title("Revenue Earned by Companies in USD_$Bn")

plt.tight_layout()
plt.show()
```

# AI Revenue Growth Year-by-Year

```python
data_openAi.head()
```

```python



data_openAi = data_openAi.copy()
data_Google = data_Google.copy()
data_Meta = data_Meta.copy()

data_openAi['Company'] = 'OpenAI'
data_Google['Company'] = 'Google'
data_Meta['Company'] = 'Meta'

# Combine all data
combined_df = pd.concat([data_openAi, data_Google, data_Meta])

# Plot grouped bar chart
plt.figure(figsize=(10, 6))
sns.barplot(
    x='Year',
    y='AI_Revenue_Growth_%',
    hue='Company',
    data=combined_df,
    palette='Set2'
)

plt.title("AI Revenue Growth Year-by-Year (OpenAI, Google, Meta)")
plt.xlabel('Year')
plt.ylabel('AI Revenue Growth (%)')
plt.legend(title='Company')
plt.show()
```

# Visualization 


## 1.1. Line Plot â€“ US Trade Deficit with Each Country

This line plot visualizes the trade deficit that the United States has with its top trading partners. A trade deficit occurs when a country's imports exceed its exports with another country, resulting in a negative balance. Here, countries on the X-axis are ranked based on their trade deficit values, and the Y-axis (inverted) shows the magnitude of deficit in USD.

To make the visualization more intuitive, we have **inverted the Y-axis**, so that **larger deficits appear higher** on the chart. This helps highlight countries with whom the US has the highest negative trade balance.


```python
# country with which us has most trade deficite


plt.figure(figsize=(12,6))
sns.lineplot(x='Country', y='US_Trade_Deficit_with_Country', data=df, marker='o')
plt.xticks(rotation=90)
plt.title('US Trade Deficit with Each Country')
plt.xlabel('Country')
plt.ylabel('Trade Deficit (in USD)')

# Reverse the Y-axis so negative values go up
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

From this visualization, we can identify the countries with the **largest trade deficits** with the United States. Countries like **China**, **Mexico**, and **Vietnam** stand out with significantly higher deficits, indicating a substantial imbalance in trade flows. This insight is crucial for understanding the **economic dependencies and policy impacts**, especially in the context of tariff impositions, international trade negotiations, and reshoring strategies.


## 1.2. Bar Plot â€“ Tariff Imposed (%) by US

This bar plot illustrates the percentage of tariffs imposed by the **United States** on imports from various countries. Each bar represents a different country, and the height of the bar reflects the **tariff rate** imposed.

This visualization helps identify countries that have been targeted with **higher trade barriers**, possibly due to trade disputes, policy shifts, or efforts to protect domestic industries. Countries with higher tariffs are likely to experience **greater disruptions** in their export economy and bilateral trade relations.


```python
# top 10 country where trump has imposed tarrif on 
df1=df.sort_values('Tarrif_imposed_%' , ascending=False , ignore_index= True ).head(10)
sns.set_style("whitegrid")
plt.figure(figsize=(18, 8))
sns.barplot(x='Country' , y='Tarrif_imposed_%' , data=df1 , edgecolor='black' )
plt.xticks(rotation=45, ha='right')
plt.title('Tariff Imposed (%) by US ', fontsize=16, fontweight='bold')
plt.xlabel('Country', fontsize=12 , fontweight='bold')
plt.ylabel('Tariff Imposed (%)', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 1.3 Bar Plot â€“ Export Loss by Country

This bar plot visualizes the export loss experienced by the top 20 countries due to tariffs imposed by the US. Export loss represents the estimated reduction in export value, showing which countries are most economically affected by these tariffs. Understanding this helps identify vulnerable economies and sectors impacted by trade policies.


```python
#export loss by country 
sns.barplot(x='Country', y='export_loss', data=df.sort_values(by='export_loss', ascending=False).head(20) , edgecolor='black')
plt.title("Export Loss by Country million($)" ,fontsize=16, fontweight='bold')
plt.xlabel('Country', fontsize=12 , fontweight='bold')
plt.ylabel('Export loss in Million $', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

From the plot, we can observe that countries like China, Canada, and Mexico face the highest export losses, indicating their significant trade exposure to the US market. The variation in export loss also reflects each country's dependency on US imports and the severity of tariffs imposed. This insight is crucial for policymakers and businesses to strategize around trade disruptions.


## 1.4 Bar Plot â€“ Export to US as % of GDP

This line plot shows the proportion of each countryâ€™s GDP that comes from exports to the United States. It highlights how dependent countriesâ€™ economies are on trade with the US, giving insight into potential vulnerabilities if tariffs affect these exports.


```python
sns.lineplot(x='Country', y='Export_to_US_as_%_of_GDP', data=df.sort_values(by='Export_to_US_as_%_of_GDP', ascending=False).head(20) , marker='o' )
plt.title("Export to US as % of GDP" ,fontsize=16, fontweight='bold')
plt.xlabel('Country', fontsize=12 , fontweight='bold')
plt.ylabel('% of Gdp which comes from us export', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

From this visualization, we see that some countries rely heavily on exports to the US relative to the size of their economy. Countries with higher export-to-US percentages might face greater economic risks due to tariff changes, emphasizing the importance of trade diversification.


## 1.5 Pie Chart â€“ Top Sectors Affected (Global Count)

This pie chart represents the distribution of the top sectors globally affected by the US tariffs. It helps us understand which industries are bearing the brunt of the trade restrictions and how the impact is spread across different sectors worldwide.


```python
sector_counts = df['Top_Sectors_Affected'].value_counts()
top_sectors = sector_counts[:5]
others = sector_counts[5:].sum()
top_sectors['Others'] = others
top_sectors.plot.pie(autopct='%1.1f%%', startangle=140, figsize=(8,8), shadow=True)
plt.ylabel('')
plt.title('Top 5 Sectors Affected by US Tariffs (Others grouped)')
plt.show()
```

From the chart, we observe which sectors face the most significant tariff impact globally. This insight highlights key areas where supply chains and businesses might be disrupted, signaling sectors that may need to strategize for mitigating these trade challenges.

# 2.1 Scatter Plot: Tariff Imposed vs Export Loss
This scatter plot shows the relationship between the tariff percentage imposed by the US on each country and the corresponding export loss that country faces. We expect to see if higher tariffs directly correlate with larger export losses. Each dot represents a country, highlighting how tariffs impact their trade with the US.

```python
import plotly.express as px

fig = px.scatter(
     df1 , x='Tarrif_imposed_%' , y='export_loss',
     size='GDP_2024 (Million USD)',
    color='export_loss',
    size_max=40,
    color_continuous_scale='Viridis'
)
fig.update_layout(title='Tariff Imposed (%) vs Export Loss by Country')
fig.show()

```

The scatter plot indicates a positive correlation between tariff rates and export loss for most countries. However, the impact varies depending on each countryâ€™s export volume to the US. Some countries with lower tariffs still experience significant losses due to their higher trade dependency, while others with higher tariffs show moderate losses.

## 2.3. Bar Plot â€“ How Export to US Shrunk (Before vs After Tariff)

This grouped bar plot visualizes the **impact of US-imposed tariffs** on exports from different countries. It compares the export values to the US **before and after tariffs** were applied.

Each pair of bars per country represents:
- **Export_to_US**: The value of exports to the US before tariffs.
- **Export_to_US_Post_Tariff**: The reduced export value after the tariffs were implemented.

This visualization clearly shows the **magnitude of export shrinkage**, helping us understand which countries have been most **economically affected**.


```python
df_melted = df5[['Country', 'export_to _us', 'Export_to_US_Post_Tarrif']].melt(id_vars='Country', var_name='Stage', value_name='Export')
sns.barplot(x='Country', y='Export', hue='Stage', data=df_melted)
plt.title('How Export to Us Shrunk ' , fontsize=20, fontweight='bold')
plt.ylabel('Export in Million $ ' , fontsize=14, fontweight='bold')
plt.xlabel('Country ' , fontsize=14, fontweight='bold')
plt.show()
```

From the chart, itâ€™s evident that countries such as **China**, **Germany**, and **Mexico** experienced a noticeable drop in their export volume to the US after tariffs were enforced.

This decline reflects the **direct economic consequences of protectionist trade measures**, potentially leading to **slower GDP growth**, **loss of trade surplus**, and **increased trade tensions**.

Such insights are essential for evaluating **policy effectiveness** and its ripple effects on **global trade networks**.


# 3.2. Word Cloud â€“ Most Affected Sectors Globally

This word cloud highlights the **most frequently impacted sectors** due to US-imposed tariffs. The text is derived from the `Top_Sectors_Affected` column in the dataset, aggregated across all countries.

```python
from wordcloud import WordCloud
text = " ".join(df['Top_Sectors_Affected'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Affected Sectors Globally')
plt.show()
```

## 3.4 ðŸ“‰ Export to US â€“ Before vs After Tariff Imposition

This grouped bar chart illustrates how exports to the US from different countries have changed due to the tariffs imposed. 

The **left bars represent exports before tariffs**, and the **right bars show exports after tariffs**. 

This visualization helps us understand the impact of US tariff policies on global trade flows.


```python
import pandas as pd
import plotly.express as px

# Make sure Tariff is numeric (remove % sign if needed)
df['Tarrif_imposed_%'] = df['Tarrif_imposed_%'].replace('%', '', regex=True).astype(float)

# Build the choropleth map
fig = px.choropleth(
    df,
    locations='Country',
    locationmode='country names',
    color='Tarrif_imposed_%',
    hover_name='Country',
    color_continuous_scale='Turbo',  # Dynamic & vibrant
    title='ðŸŒ Global Tariff Impact â€“ Tariff Imposed (%) by Country',
)

# Update layout for better appearance
fig.update_layout(
    title_font_size=24,
    title_x=0.5,
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type='natural earth'
    ),
    height=700,   # Larger height
    width=1200,   # Wider width
    margin=dict(l=20, r=20, t=60, b=20)
)

# Show map
fig.show()

```
